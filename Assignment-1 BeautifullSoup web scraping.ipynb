5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
i) Headline
ii) Time
iii) News Link

```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.cnbc.com/world/?region=world'
response = requests.get(url)
```

```python
soup=BeautifulSoup(response.text,'html.parser')
```

```python
headline = soup.find('div',class_="LatestNews-headlineWrapper")
headline
```

```python
headline.text
```

```python
headline=[]

for i in soup.find_all('div',class_="LatestNews-headlineWrapper"):
     headline.append(i.text)

headline
```

```python
link=soup.find('li',class_="LatestNews-item")
link               
```

```python
link.text
```

```python
link=[]
for i in soup.find_all('li',class_="LatestNews-item"):
    link.append(i.text)
link    
```

```python
time= soup.find("span",class_="LatestNews-wrapper")
time
```

```python
time.text
```

```python
time =[]
for i in soup.find_all("span",class_="LatestNews-wrapper"):
    
  time.append(i.text)
                       
                      
time                       
                    
```

```python
#printing length
print(len(link),len(headline),len(time))
```

```python
import pandas as pd

df= pd.DataFrame({'Link':link,'Headline':headline,'Time':time})

df
```

7) Write a python program to scrape mentioned details from dineout.co.in and make data frame-

```python
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
Restro_Name=soup.find('div',class_="restnt-info cursor")
Restro_Name
```

```python
Restro_Name.text
```

```python
Restro_Name=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    Restro_Name.append(i.text)
Restro_Name    
```

```python
loc=soup.find('div',class_="restnt-loc ellipsis")
loc
```

```python
loc.text
```

```python
loc=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    loc.append(i.text)
loc    
```

```python
price= soup.find('span',class_="double-line-ellipsis")
price
```

```python
price.text
```

```python
price=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text)
price    
```

```python
image= soup.find('img',class_="no-img")
image
```

```python
image=[]
for i in soup.find_all('img',class_="no-img"):
    image.append(i['data-src'])
image    
    
```

```python
rating =[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
rating    
```

```python
import pandas as pd
df=pd.DataFrame({'Restaurant':Restro_Name,'Location':loc,'Cusine':price,'Images':image,'Rating':rating})
df
```
