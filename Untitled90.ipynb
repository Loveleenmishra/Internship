# World happiness score

Importing important laibries

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('darkgrid')
plt.rcParams['font.size']=15
plt.rcParams['figure.figsize']=(10,7)
plt.rcParams['figure.facecolor']='#FFE5b4'
```

```python
data=pd.read_csv('happiness_score_dataset.csv')
```

```python
data.head()
```

```python
data_coloumns=['Country','Region','Happiness Score','Economy (GDP per Capita)','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity']
```

```python
data=data[data_coloumns].copy()
```

```python
happy_df= data.rename({'Country':'country_name','Region':'regional_indicator',"Happiness Score":'happiness_score','Economy (GDP per Capita)':'logged_GDP_per_capita','Health (Life Expectancy)':'healthy_life_expectacy','Freedom':'freedom_to_make_life_choises','Trust (Government Corruption)':'perceptions_of_corruption','Generosity':'generosity' }, axis=1)
```

```python
happy_df.head()
```

Checking null value

```python
happy_df.isnull().sum()
```

Visualisation

we create plot between happiness score and gdp for diffrent region

```python
plt.rcParams['figure.figsize']=(15,7)
plt.title('Plot between Happiness Score and GDP')
sns.scatterplot(x=happy_df.happiness_score, y=happy_df.logged_GDP_per_capita, hue= happy_df.regional_indicator,s=200);

plt.legend(loc='upper left', fontsize='10')
plt.xlabel('Happiness Score')
plt.ylabel('GDP per capita')
```

```python
gdp_region =  happy_df.groupby('regional_indicator')['logged_GDP_per_capita'].sum()
gdp_region
```

```python
#Total countries
total_country= happy_df.groupby('regional_indicator')[['country_name']].count()
print(total_country)
```

```python
#correlation map

cor = happy_df.corr(method = "pearson")
f,ax= plt.subplots(figsize=(10,5))
sns.heatmap(cor,mask =np.zeros_like(cor, dtype=np.bool),
            cmap="Blues", square=True, ax=ax)
```

Here above we have correlation matrix, here my C map is blue in color.
wherever the cells are in blue and dark blue color it's mean the variables are very high.
whereever you see the light blue,white,grey color they indicate that this varible are very low in corelation  


```python
#corruption in region

corruption= happy_df.groupby('regional_indicator')[['perceptions_of_corruption']].mean()
corruption
```

```python
plt.rcParams['figure.figsize']=(12,8)
plt.title('Perception of Corruption in various Regions')
plt.xlabel('Regions',fontsize = 15)
plt.ylabel('Corruption Index',fontsize= 15)
plt.xticks(rotation=30, ha='right')
plt.bar(corruption.index,corruption.perceptions_of_corruption)
```

In above graph we see least corruption in central and Eastern europe & highest corruption in  australia and new zealand. 

```python
top_10=happy_df.head(10)
bottom_10 = happy_df.tail(10)
```

```python
fig, axes= plt.subplots(1,2, figsize=(16,6))
plt.tight_layout(pad=2)
xlabels= top_10.country_name
axes[0].set_title('TOP 10 happiest countries Life Expectancy')
axes[0].set_xticklabels(xlabels, rotation=45, ha='right')
sns.barplot(x=top_10.country_name, y= top_10.healthy_life_expectancy, ax= axes[0])
axes[0].set_xlabel('Country Name')
axes[0].set_ylabel('Life expectancy')

xlabels= Bottom_10.country_name
axes[1].set_title('Bottom 10 least happy countries Life Expectancy')
axes[1].set_xticklabels(xlabels, rotation=45, ha='right')
sns.barplot(x=bottom_10.country_name, y= bottom_10.healthy_life_expectancy, ax= axes[1])
axes[1].set_xlabel('Country Name')
axes[1].set_ylabel('Life expectancy')
```

```python
country = happy_df.sort_values(by= 'perceptions_of_corruption').head(10)
plt.rcParams['figure.figsize']=(12,6)
plt.title('Countries with most Perception of Corruption')
plt.xlabel('Countries',fontsize = 13)
plt.ylabel('Corruption Index',fontsize= 13)
plt.xticks(rotation=30, ha='right')
plt.bar(country.country_name,country.perceptions_of_corruption)
```

These are countries which is least perception of corruption.(this is from head)

```python
country = happy_df.sort_values(by= 'perceptions_of_corruption').tail(10)
plt.rcParams['figure.figsize']=(12,6)
plt.title('Countries with most Perception of Corruption')
plt.xlabel('Countries',fontsize = 13)
plt.ylabel('Corruption Index',fontsize= 13)
plt.xticks(rotation=30, ha='right')
plt.bar(country.country_name,country.perceptions_of_corruption)
```
